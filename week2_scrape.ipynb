```python
week2_link = 'http://www.footballlocks.com/nfl_odds_week_2.shtml#Closing%20NFL%20Odds%20Week%201,%202007'

#get html text
week2_html = requests.get(week2_link).text

soup = BeautifulSoup(week2_html, 'html.parser')

#want just table row data
tr = soup.find_all("tr")

len(tr)

#start of table
print(tr[51])

#last entry in table
print(tr[295])
```

```python
#want to 296 because it does not include last one
table = tr[51:296]

type(table[-1])

type(table)

string_table = [str(i) for i in table]

type(string_table[0])

week2_df = pd.DataFrame(string_table)

week2_df.columns = ['all_data']

week2_df.head()
```

```python
#want to clean the date data

dates = [x.split('td')[1] for x in week2_df['all_data']]

cleaned_dates = []
for x in dates:
    str1 = x.replace(">", "").replace("<", "")
    final_str = str1[:-1]
    cleaned_dates.append(final_str)

week2_df['date'] = cleaned_dates
```

```python
#pull out favorites

week2_df['cleaned_favorites'] = week2_df['all_data'].map(lambda x: x.split('td')[3].replace('<', '').replace('>', '').replace('/', '')
                                                 if len(x.split('td'))>3 else np.nan)

#create underdog column

week2_df['cleaned_underdogs'] = week2_df['all_data'].map(lambda x: x.split('td')[7].replace('<', '').replace('>', '').replace('/', '')
                                                 if len(x.split('td'))>3 else np.nan)

#create spread column
week2_df['cleaned_spread'] = week2_df['all_data'].map(lambda x: x.split('td')[5].replace('<', '').replace('>', '').replace('/', '')
                                                 if len(x.split('td'))>3 else np.nan)

#create over_under column
week2_df['cleaned_over_under'] = week2_df['all_data'].map(lambda x: x.split('td')[9].replace('<', '').replace('>', '').replace('/', '')
                                                 if len(x.split('td'))>3 else np.nan)

#create money_odds column and clean data
money_odds = []
for x in week2_df['all_data']:
    try:
        money_odds.append(x.split('td')[11].replace('<', '').replace('>', '').replace('/', ''))
    except:
        money_odds.append(np.nan)
        
week2_df['money_odds'] = money_odds

#input week number for each game
week2_df['week_number'] = '2'
```

```python
#want to input year for each game. on lines we want to decrease the count by one, len(week1_df.all_data[x].split('td')) = 3
year = 2018
year_list = []
for x in week2_df['all_data']:
    if len(x.split('td')) > 3:
        year_list.append(year)
    else:
        year_list.append(year)
        year -= 1

week2_df['year'] = year_list
```

```python
#create a final dataframe for week 1
week2_final = week2_df[['date', 'year', 'week_number', 'cleaned_favorites', 'cleaned_underdogs', 'cleaned_spread', 
                        'cleaned_over_under', 'money_odds']]
week2_final.head()
```
